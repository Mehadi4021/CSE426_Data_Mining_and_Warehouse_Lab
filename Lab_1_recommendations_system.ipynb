```python
import pandas as pd
```

```python
users = ['Sajid', 'Arbin', 'Sohana', 'Santo', 'Rimon', 'Tanvir', 'Saad', 'Fudail', 'Tawhid', 'Rian']

ratings = {
    'Pushpa': [3.5, 2, 3, 2, 3, 0, 0, 1.5, 3, 3],
    '3 Idiots': [4.5, 5, 5, 5, 5, 5, 5, 5, 4, 5],
    'Dilwale': [4, 4.5, 4.8, 0, 4.5, 0, 4.7, 4.5, 5, 4],
    'Endgame': [4.7, 4.5, 3.5, 4.8, 4, 5, 4.9, 5, 5, 5],
    'KGF': [3.7, 4, 4, 3, 3.5, 0, 4.2, 5, 4, 4]
}
```

Recommendation Movie

```python
for movie in ratings.keys():
  r_sum =sum(ratings[movie])
  r_n = len([i for i in ratings[movie] if i!=0])
  print(movie, r_sum/r_n)
```

```python
ratings_df = pd.DataFrame(ratings, index=users)
ratings_df
```

```python
from scipy.spatial.distance import euclidean

def euclidean_similarity(user1, user2):
  distance = euclidean(user1, user2)
  similarity = 1/(distance+1)
  return similarity
```

```python
euclidean_similarity(ratings_df.loc['Tanvir'], ratings_df.loc['Saad'])
```

```python
target_user = 'Tawhid'

user_similarities = {}
```

```python
for i in users:
  if i != target_user:
    user_similarities[i]= euclidean_similarity(ratings_df.loc[target_user], ratings_df.loc[i])
print(user_similarities)
```
